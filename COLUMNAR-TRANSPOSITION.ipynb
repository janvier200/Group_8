## Columnar Transposition Columnar <br><br>Rearranges plaintext by writing in rows and reading columns in a key‑determined order.

## Encryption

```python
def encrypt_columnar(plaintext, key):
    # Remove spaces from plaintext and convert to uppercase
    plaintext = plaintext.replace(" ", "").upper()
    
    # Convert key to uppercase
    key = key.upper()
    
    # Number of columns = length of the key
    col = len(key)
    
    # Calculate number of rows needed (ceiling division)
    row = (len(plaintext) + col - 1) // col
    
    # Create an empty matrix (list of lists) with rows and columns
    matrix = [['' for _ in range(col)] for _ in range(row)]
    
    # Fill the matrix row by row with plaintext characters
    index = 0
    for r in range(row):
        for c in range(col):
            if index < len(plaintext):
                matrix[r][c] = plaintext[index]  # place character in matrix
                index += 1
            else:
                matrix[r][c] = 'X'  # add padding if plaintext ends
    
    # Sort the key while keeping track of original column indices
    # enumerate(key) gives (index, letter)
    key_order = sorted(list(enumerate(key)), key=lambda x: x[1])
    
    # Read columns in the order determined by sorted key
    ciphertext = ""
    for k in key_order:
        col_index = k[0]  # original column position
        for r in range(row):
            ciphertext += matrix[r][col_index]  # read column top to bottom
    
    # Return the final encrypted message
    return ciphertext
```

## Decryption

```python
def decrypt_columnar(ciphertext, key):
    # Convert key to uppercase to match encryption process
    key = key.upper()
    
    # Number of columns = length of the key
    col = len(key)
    
    # Number of rows = total characters divided by columns
    row = len(ciphertext) // col
    
    # Sort the key while keeping original indices
    # This gives the order in which columns were read during encryption
    key_order = sorted(list(enumerate(key)), key=lambda x: x[1])
    
    # Create an empty matrix with given rows and columns
    matrix = [['' for _ in range(col)] for _ in range(row)]
    
    # Fill the matrix column by column based on sorted key order
    index = 0
    for k in key_order:
        col_index = k[0]  # original position of the column
        for r in range(row):
            matrix[r][col_index] = ciphertext[index]  # place character
            index += 1
    
    # Read the matrix row by row to reconstruct plaintext
    plaintext = ""
    for r in range(row):
        for c in range(col):
            plaintext += matrix[r][c]
    
    # Remove padding 'X' characters added during encryption
    return plaintext.rstrip('X')
```

```python
# Define the key used for encryption and decryption
key = "masters"

# Original message (plaintext)
message = "kill corona virus at 11 am tomorrow"

# Call the encryption function to convert plaintext into ciphertext
cipher = encrypt_columnar(message, key)

# Display the encrypted message
print("Encrypted:", cipher)

# Call the decryption function to convert ciphertext back to plaintext
decrypted = decrypt_columnar(cipher, key)

# Display the decrypted message (should match original message without spaces and in uppercase)
print("Decrypted:", decrypted)
```
