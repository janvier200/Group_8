## Columnar Transposition Columnar <br><br>Rearranges plaintext by writing in rows and reading columns in a key‑determined order.

## Encryption

```python
def encrypt_columnar(plaintext, key):
    plaintext = plaintext.replace(" ", "").upper()
    key = key.upper()
    col = len(key)
    row = (len(plaintext) + col - 1) // col
    
    matrix = [['' for _ in range(col)] for _ in range(row)]
    
    index = 0
    for r in range(row):
        for c in range(col):
            if index < len(plaintext):
                matrix[r][c] = plaintext[index]
                index += 1
            else:
                matrix[r][c] = 'X'  # padding
    
    # Sort key
    key_order = sorted(list(enumerate(key)), key=lambda x: x[1])
    
    # Read columns based on sorted key
    ciphertext = ""
    for k in key_order:
        col_index = k[0]
        for r in range(row):
            ciphertext += matrix[r][col_index]
    
    return ciphertext
```

## Decryption

```python
def decrypt_columnar(ciphertext, key):
    key = key.upper()
    col = len(key)
    row = len(ciphertext) // col
    
    # Sort key
    key_order = sorted(list(enumerate(key)), key=lambda x: x[1])
    
    # Create empty matrix
    matrix = [['' for _ in range(col)] for _ in range(row)]
    
    index = 0
    for k in key_order:
        col_index = k[0]
        for r in range(row):
            matrix[r][col_index] = ciphertext[index]
            index += 1
    
    # Read row-wise
    plaintext = ""
    for r in range(row):
        for c in range(col):
            plaintext += matrix[r][c]
    
    return plaintext.rstrip('X')  # remove padding
```

```python
key = "masters"
message = "kill corona virus at 11 am tomorrow"

cipher = encrypt_columnar(message, key)
print("Encrypted:", cipher)

decrypted = decrypt_columnar(cipher, key)
print("Decrypted:", decrypted)
```

## SHA / SHA‑1 <BR> <BR> ### Secure Hash Algorithm (160‑bit hash, developed by NIST).

```python
import hashlib

text = "hell"
hash_object = hashlib.sha1(text.encode())
print(hash_object.hexdigest())
```
